```python
train_df = pd.read_csv('data/train.csv')
predict_df = pd.read_csv('data/test.csv')


display(train_df)
```

```python
#data processing
print(train_df.info())
```

```python
train_df['Age_interv']=pd.cut(train_df['Age'],20)
predict_df['Age_interv']=pd.cut(predict_df['Age'],20)

train_df=pd.get_dummies(train_df,columns=['Sex','Embarked', 'Age_interv', 'SibSp','Parch','Ticket'])
predict_df=pd.get_dummies(predict_df,columns=['Sex','Embarked', 'Age_interv', 'SibSp','Parch','Ticket'])

y=train_df['Survived'].values
train_df=train_df.drop(['Fare','PassengerId', 'Survived','Name','Cabin','Age'], axis=1)
predict_df=predict_df.drop(['Fare','PassengerId','Name','Cabin','Age'], axis=1)
```

```python

display(train_df)
X = train_df.values
X_predict=predict_df.values
print(X)


```

```python
print(train_df.info())
```

```python
print(predict_df.info())
```

```python
#start learning
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=.1, random_state=42)
classifiers = [LogisticRegression(tol=0.000001),
               KNeighborsClassifier(),
               SVC(kernel="linear", C=0.025),
               SVC(gamma=2, C=1),
               GaussianProcessClassifier(),
               DecisionTreeClassifier(),
               RandomForestClassifier(),
               MLPClassifier(),
               MLPClassifier(hidden_layer_sizes=(100,100)),
               AdaBoostClassifier(),
               GaussianNB(),
               QuadraticDiscriminantAnalysis()]
res=[]
mxscore=0
for clf in classifiers:
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print(score)
    print(clf)
    if score>mxscore: 
        optimclf=clf
        mxscore=score
    res.append((score,clf))
    
```

```python
out=optimclf.predict(X_predict)
dfout=pd.DataFrame(data={'PassengerId': predict_df['PassengerId'].values, 'Survived': out})
dfout.to_csv('out.csv',index=False)
```
